# Avocado Market Prediction

- The objective of this case study is to predict the price of avocado sold in US using historical dataset.
- Dta repreesents weekly 2018 retail scanned data for the National retail volume (units) and price.
- The data is as follows:


- ```Date - the date of the observation```
- ```Average Price - the average price of a single avocado ```
- ```Type - conventional or organic```
- ```Year - the year```
- ```Region - the city or region of the observation```
- ```Total Volume - total number of avocados sold```
- ```4046 - total number of avocados PLU 4046 sold```
- ```4225 - total number of avocados PLU 4225 sold```
- ```4770 - total number of avocados PLU 4770 sold```

*PLU - Product Lookup Codes*

## What is Facebook Prophet?

- Prophet is open source software released by Facebook's Core Data Science team.
- Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly,weekly and daily seasonality, plus holiday effects.
- Prophet works best with time series that have strong seasonal effects and several seasons of historical data.
- Prophet implements an additive regression model with four elements:
    - A piecewise linear, prophet automatically picks up change points in the data and indentifies any change in trends.
    - A yearly seasonal component modeled using Fourier series.
    - A weekly seasonal componen 
    - A holiday list that can be manually provided.
   

Additive Regression model takes the form:
    ```Y=β_0+∑_j=1^p f_j(X_j)+ϵ ```
 The functions f(x) are unknown smoothing functions fit from the data

````python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']] 
````

````python
avocado_prophet_df
````

## Data Preparation For Prophet

````python
avocado_prophet_df = avocado_prophet_df.rename(columns={'Date':'ds', 'AveragePrice':'y'})
````

````python
avocado_prophet_df
````

## Creating Prophet

````python
m = Prophet()
m.fit(avocado_prophet_df)
````

## Forecasting into Future

````python
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
````

````python
forecast
````

````python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
````

````python
figure3 = m.plot_components(forecast)
````
